```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Project_NLP/vnexpress_articles_merge.csv')
print(df['full_text'])
```

```python
from underthesea import sent_tokenize
from underthesea import word_tokenize

# Giả sử bạn đã có DataFrame df với cột 'full_text'

# Đếm số câu bằng underthesea
df['sentence_count'] = df['full_text'].apply(lambda x: len(sent_tokenize(str(x))))
df['sumary_token']= df['summary'].apply(lambda x: len(word_tokenize(str(x))))
```

```python
# Lọc ra các dòng có nhiều hơn 1 câu
df_filtered = df[df['sentence_count'] > 2].copy()

```
